```python
import pandas as pd
from scipy.stats import entropy
from sklearn import tree
import matplotlib
import sklearn
```

```python
df = pd.read_csv('cats.csv')
df
```

```python
X = df[['Шерстист']]
y = df['Вид']
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
clf.fit(X,y)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf, filled=True)
```

```python
X = df[['Гавкает']]
clf.fit(X,y)
plot_tree(clf, filled=True)
```
