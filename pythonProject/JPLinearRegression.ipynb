```python
from pydataset import data
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
%matplotlib notebook
```

```python
pima = data('Pima.tr')
pima
```

```python
pima.plot(kind='scatter', x='skin', y='bmi')
plt.show()
```

```python
#test train split for supervised traning
X_train, X_test, y_train, y_test = train_test_split(pima.skin, pima.bmi ,test_size=0.33,random_state=42)
```

```python
#test train split plot
plt.scatter(X_train, y_train, label='Traning data', color='r', alpha=.7)
plt.scatter(X_test, y_test, label='Testing data', color='g', alpha=.7)
plt.legend()
plt.title("Test Train Split")
```
